```python
import matplotlib.pyplot as plt
import polars as pl

bigDf = pl.scan_csv("/Users/varunsingh/Desktop/Projects/ResolveIQ/ResolveIQ/docs/exploration/data/complaints.csv")

t_years = [2023, 2024, 2025]

df = bigDf.filter(
    pl.any_horizontal([pl.col('Date received').str.starts_with(str(year)) for year in t_years])
)

print(f"Shape: {df.select(pl.len()).collect().item()}")
print(f"Columns: {df.collect_schema().names()}")
```

Columns selected here will form a foundational basis for:
1. ML Model selection
2. Feature Engineering
3. RAG retrieval and agent direction

The below cells are an analysis of determining key metrics!

```python
# Distinct Products

distProd = (
    df.with_columns(
        pl.col('Product')
    )
    .group_by('Product')
    .agg(pl.len().alias('Pcount'))
    .sort('Pcount', descending=True)
    .collect()
)

print(f"Distinct Product: {distProd}")
```

```python
# Distinct Sub P cnt

distSubProd = (
    df.with_columns(
        pl.col('Sub-product')
    )
    .group_by('Sub-product')
    .agg(pl.len().alias('Sub-productCnt'))
    .sort('Sub-productCnt', descending=True)
    .collect()
)

print(f"Distinct Sub P: {distSubProd}")
```

```python
# Distinct Issue

issue = (
    df.with_columns(
        pl.col('Issue')
    )
    .group_by('Issue')
    .agg(pl.len().alias('IssueCnt'))
    .sort('IssueCnt', descending=True)
    .collect()
)

print(f"Distinct Issue: {issue}")
```

```python
# Distinct state

state = (
    df.with_columns(
        pl.col('State')
    )
    .group_by('State')
    .agg(pl.len().alias('StateCnt'))
    .sort('StateCnt', descending=True)
    .collect()
)

print(f"Distinct State: {state}")
```

```python
# Distinct Company

company = (
    df.with_columns(
        pl.col('Company')
    )
    .group_by('Company')
    .agg(pl.len().alias('CompanyCnt'))
    .sort('CompanyCnt', descending=True)
    .collect()
)

print(f"Distinct Company: {company}")
```

```python
# Value counts for:

counts_df = df.select(pl.col('Submitted via').value_counts(sort=True)).unnest('Submitted via').collect()
print(counts_df)
```

```python
# The responses received by the customer:

resp_df = df.select(pl.col('Company response to consumer').value_counts(sort=True)).unnest('Company response to consumer').collect()
print(resp_df)
```

```python
# 1. Group, count, and safely calculate percentages
normalized_counts = (
    df.group_by("Timely response?")
    .agg(pl.len().alias("Count")) # First, count rows per group
    .with_columns(
        # Then, divide each group count by the total sum of all counts
        ((pl.col("Count") / pl.col("Count").sum()) * 100).round(2).alias("Percentage (%)")
    )
    .sort("Percentage (%)", descending=True)
)

# 2. Trigger execution if 'df' is your 8GB LazyFrame
if isinstance(normalized_counts, pl.LazyFrame):
    normalized_counts = normalized_counts.collect()

print(normalized_counts)
```

```python
# 1. Group, count, and safely calculate percentages
normalized_counts = (
    df.group_by("Consumer complaint narrative")
    .agg(pl.len().alias("Count")) # First, count rows per group
    .with_columns(
        # Then, divide each group count by the total sum of all counts
        ((pl.col("Count") / pl.col("Count").sum()) * 100).round(2).alias("Percentage (%)")
    )
    .sort("Percentage (%)", descending=True)
)

# 2. Trigger execution if 'df' is your 8GB LazyFrame
if isinstance(normalized_counts, pl.LazyFrame):
    normalized_counts = normalized_counts.collect()

print(normalized_counts)
```

```python
narrative_col = "Consumer complaint narrative"

# 1. Filter out nulls and calculate lengths efficiently on the 8GB file
lengths_df = (
    df.filter(pl.col(narrative_col).is_not_null())  # Drops rows without narratives
    .select(
        pl.col(narrative_col).str.len_chars().alias("narrative_len") # Calculates text length
    )
    .collect()  # Safely triggers the execution and loads ONLY the lengths into RAM
)

print(lengths_df["narrative_len"].describe())

# 3. Pull the column into a list and apply the clip logic for plotting
clipped_lens = lengths_df["narrative_len"].clip(upper_bound=5000).to_list()

# 4. Generate the visualization
plt.figure(figsize=(10, 4))
plt.hist(clipped_lens, bins=50, color="#1f77b4", edgecolor="black", alpha=0.7)
plt.xlabel("Narrative length (characters)")
plt.ylabel("Count")
plt.title("Distribution of complaint narrative lengths")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.show()
```
